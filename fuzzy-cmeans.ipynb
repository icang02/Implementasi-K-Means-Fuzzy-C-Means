```python
from fcmeans import FCM
import pandas as pd
import matplotlib.pyplot as plt


def read_data(file_path):
    file = pd.read_csv(file_path)
    return file.values


def fcmeans_clustering(dataset, k):
    # Inisialisasi dan fitting model FCM
    my_model = FCM(n_clusters=k)
    my_model.fit(dataset)

    # Menentukan pusat cluster dan label
    centers = my_model.centers
    labels = my_model.predict(dataset)

    # Variabel warna untuk tiap cluster
    cluster_colors = [
        "red",
        "orange",
        "blue",
        "green",
        "purple",
        "brown",
        "pink",
        "gray",
        "olive",
        "cyan",
    ]  # 10 color

    # Visualisasi grafik clustering
    plt.scatter(
        dataset[:, 0],
        dataset[:, 1],
        c=[cluster_colors[label] for label in labels],
        alpha=0.5,
    )
    plt.scatter(
        centers[:, 0], centers[:, 1], c="black", marker="X", s=100, label="Centroids"
    )
    plt.title("FCM Clustering â€” Gizi Balita")

    for i in range(k):
        plt.scatter([], [], c=cluster_colors[i], label=f"Cluster {i+1}")

    plt.legend()
    plt.show()


# Main Program
print("==FCM Clustering==")
print("1. Load data")
print("2. Pilih jumlah cluster")
print("3. Run data")
print("4. Keluar")

while True:
    p = int(input("Masukkan pilihan : "))

    if p == 1:
        file_path = input("Masukkan nama file : ")
        dataset = read_data(file_path)

    elif p == 2:
        k = int(input("Masukkan jumlah cluster : "))

    elif p == 3:
        fcmeans_clustering(dataset, k)

    elif p == 4:
        break

    else:
        print("Menu tidak tersedia.")
```
